### Probability Event:
- Para cada evento $E$, a probabilidade de um evento é representada por:
\begin{equation}
P(E)
\end{equation}
<br/>
<br/>

- Fórmula:
\begin{align}
P(E) =  \frac{elementos(E)}{elementos(\Omega)}	
\end{align}
<br/>
<br/>

- Toda probabilidade está entre 0 e 1 incluído:<br/>
\begin{equation}\label
 00⩽P(E)⩽1
\end{equation}
<br/>
<br/>

### Probabily in Python

```python
# Create function that returns probability percent rounded to one decimal place
def event_probability(event_outcomes, sample_space):
    probability = (event_outcomes / sample_space) * 100
    return round(probability, 1)
```

```python
# Sample Space
cards = 52

# Determine the probability of drawing a heart
hearts = 13
heart_probability = event_probability(hearts, cards)

# Determine the probability of drawing a face card
face_cards = 12
face_card_probability = event_probability(face_cards, cards)

# Determine the probability of drawing the queen of hearts
queen_of_hearts = 1
queen_of_hearts_probability = event_probability(queen_of_hearts, cards)

# Print each probability
print('Choose a heart card: '+ str(heart_probability) + '%')
print('Choose a face card: '+ str(face_card_probability) + '%')
print('Choose a queen of heart: '+ str(queen_of_hearts_probability) + '%')
```

### Sum Theorem
Sendo (A) e (B) eventos do mesmo espaço amostral $/omega$.<br/><br/>
\begin{equation*}
P (A ∪ B) = P(A) + P(B) - P(A ∩ B)
\end{equation*}


\begin{exercise}
Empregados sairam de uma empresa pelos motivos abaixo:

- 30% salários baixos 
- 20% atribuições ruins
- 12% salários baixos + atribuições ruins


Qual a probabilidade de um empregado ter saido por causado salário baixo OU atribuições ruins?<br/>
\end{exercise}


Resolução:<br/>
**1º passo - espaço amostral (S):**<br/>
$S$ = {salários baixos, atribuições ruins, (salários baixos + atribuições ruins)}

**2º passo - Eventos:**<br/>
"salário baixo OU atribuições ruins" é uma SOMA de 2 eventos.<br/>
Evento A, P(A)<br/>
Evento B, P(B)<br/>

**3º passo - Análise das condições:**<br/>
A condição é: não pode ocorrer o evento: (salários baixos + atribuições ruins).<br/>
Os eventos devem ser mutuamente excludentes (sem elementos em comum).

**4º passo - Sum Theorem:**<br/>
<table>
  <tr>
    <th>P (A ∪ B) = 0.3 + 0.2 - 0.12 = 38%</th>
  </tr>
</table>





### Independence
**Independent Events:** A and B are independent if knowing whether **(A) occurred gives no information about whether (B) occurred**.<br/><br/>
\begin{equation*}
P (A ∩ B) = P (A)P (B)
\end{equation*}
\begin{equation*}
P (A | B) = P (A)
\end{equation*}
\begin{equation*}
P (B | A) = P (B)
\end{equation*}
OBS: quando tiramos algo sem reposição modificamos o espaço amostral.

\begin{exercise}
Uma urna contem 2 bolas bancas e 3 vermelhas, são retirados 2 bolas.<br/>
- a) Se retirado SEM reposição, qual a probabilidade das bolas serem da mesma cor?<br/>
- b) Se retirado COM reposição, qual a probabilidade das bolas serem da mesma cor?<br/>

\end{exercise}


Resolução:<br/>
**1º passo - espaço amostral (S):**<br/>
2 bolas brancas + 3 bolas vermelhas = 5

**2º passo - Eventos:**<br/>
Retirada de bola<br/>
Retirada de bola<br/>

**3º passo - Análise das condições:**<br/>
A condição é: "serem da mesma cor"

**4º passo - Sem reposição no $(E)$ independentes:**<br/>
1º e 2º eventos:<br/>
\begin{equation*}
P(branca ∩ branca) = P(branca) * P(branca) = (2/5) * (1/4) = 1/10
\end{equation*}
\begin{equation*}
P(vermelha ∩ vermelha) = P(vermelha) * P(vermelha) = (3/5) * (2/4) = 3/10
\end{equation*}

Agora é necessário calcular a união destes eventos.
<table>
  <tr>
    <th>P(mesma cor) = P(vermelha ∩ vermelha) ∪ P(branca ∩ branca) = 40%</th>
  </tr>
</table>

**5º passo - Com reposição no $(E)$ dependentes:**<br/>
1º e 2º eventos:<br/>
\begin{equation*}
P(branca ∩ branca) = P(branca) * P(branca) = (2/5) * (2/5) = 4/25
\end{equation*}

\begin{equation*}
P(vermelha ∩ vermelha) = P(vermelha) * P(vermelha) = (3/5) * (3/5) = 9/25
\end{equation*}

Agora é necessário calcular a união destes eventos.
<table>
  <tr>
    <th>P(mesma cor) = P(vermelha ∩ vermelha) ∪ P(branca ∩ branca) = 52%</th>
  </tr>
</table>

### Bayes Theorem
\begin{equation*}
P (A|B) = \frac{P (B|A)P (A)}{P (B)}
\end{equation*}
