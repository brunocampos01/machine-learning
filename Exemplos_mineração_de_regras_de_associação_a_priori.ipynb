```
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

###Carrega o conjunto de dados

```
url = "https://codigos.ufsc.br/jonata.tyska/repositorio/-/raw/master/datasets/order_data.csv"

# Read dataset to pandas dataframe
dataset = pd.read_csv(url,delimiter=" ",header=None)
dataset.head()
```

### Convertendo os dados para o formato da biblioteca apyori

```
data_list = []
for row in range(0, 20):
    data_list.append(list(filter(lambda x: x != 'nan', [str(dataset.values[row,column]) for column in range(0, 9)])))

data_list
```

```
algo = apriori(data_list, min_support=0.1, min_confidence=0.1, min_lift=2, min_length=2)
results = list(algo)
```

```
for i in range(0,len(results)):
    print(f"Required Association No. {i+1} is: {results[i]}")
    print('-'*25)
```

### Exemplo usando a biblioteca mlxtend

```
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('https://codigos.ufsc.br/jonata.tyska/repositorio/-/raw/master/datasets/retail_dataset.csv', sep=',')
df.head(5)
```

```
items = (df['0'].unique())
items
```

```
itemset = set(items)
encoded_vals = []
for index, row in df.iterrows():
    rowset = set(row) 
    labels = {}
    uncommons = list(itemset - rowset)
    commons = list(itemset.intersection(rowset))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
ohe_df = pd.DataFrame(encoded_vals)
ohe_df.head()
```

### Calcula o conjunto de itens frequentes

```
freq_items = apriori(ohe_df, min_support=0.2, use_colnames=True)
freq_items
```

### Mineração das regras de associação

```
association_rules(freq_items, metric='confidence', min_threshold=0.6, support_only=False)
```
