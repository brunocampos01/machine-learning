#### Function to plot percentage on top of factor plots

```python
# Credit: Adapted from http://stackoverflow.com/a/31754317
def showPercentageFactorPlot(axes):
    for p in axes.patches:
        height = p.get_height()
        axes.text(p.get_x() + p.get_width()/2.,
                  height + 0.01,
                  '{:1.1f}%'.format(height*100), ha="center")
```

### HEATMAP

#### Correlation Features of Person 

```python
def correlation(data):
    correlation = data.corr(method='pearson')
    sns.heatmap(correlation, 
                annot=True, 
                cbar=True, 
                linecolor='white',
                cmap="RdYlGn")
```

```python
plt.title('Correlation Features', y=1.05, size=15)

correlation(df)
```

#### Correlation Features of Kendall

```python
def correlation_kendall(data):
    correlation = data.corr(method='kendall')
    sns.heatmap(correlation, 
                annot=True, 
                cbar=True, 
                linecolor='white',
                cmap="RdYlGn")
```

```python
plt.title('Correlation Features of Kendall', y=1.05, size=15)

correlation_kendall(df)
```

#### Correlation Features of Spearman

```python
def correlation_spearman(data):
    correlation = data.corr(method='spearman')
    sns.heatmap(correlation, 
                annot=True, 
                cbar=True, 
                linecolor='white',
                cmap="RdYlGn")
```

```python
plt.title('Correlation Features of Spearman', y=1.05, size=15)

correlation_spearman(df)
```
