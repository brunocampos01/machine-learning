# Example Decision Tree Regression

a árvore de decisão é usada para ajustar uma curva senoidal com adição de observação ruidosa. Como resultado, ele aprende regressões lineares locais aproximando-se da curva senoidal.


Podemos ver que, se a profundidade máxima da árvore (controlada pelo parâmetro max_depth ) for muito alta, as árvores de decisão aprendem detalhes muito pequenos dos dados de treinamento e aprendem com o ruído, ou seja, preparam um overfit.



```python
# SUPERVISED LEARNING
# import
# load
# X AND Y
# classifier or regressor
# fit
# predict
# plot
```

```python
import numpy as np
import pylab as pl

from sklearn import tree

# Create a random dataset with numbers of probability distributions (RandomState())
rng = np.random.RandomState(1) 
X = 5 * rng.rand(80, 1) # rand() cria uma matriz
X = np.sort(X, axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
print(type(rng))
# print(X)
```

```python
# classifier or regressor
regressor_1 = tree.DecisionTreeRegressor(max_depth = 2)
regressor_2 = tree.DecisionTreeRegressor(max_depth = 15)

# fit (train)
regressor_1 = regressor_1.fit(X,y)
regressor_2 = regressor_2.fit(X,y)

# Predict
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regressor_1.predict(X_test)
y_2 = regressor_2.predict(X_test)
```

```python
# Plot the results
pl.figure()
pl.scatter(X, y, c="k", label="data")
pl.plot(X_test, y_1, c="g", label="max_depth=2", linewidth=2)
pl.plot(X_test, y_2, c="r", label="max_depth=5", linewidth=2)
pl.xlabel("data")
pl.ylabel("target")
pl.title("Decision Tree Regression")
pl.legend()
pl.show()
```
